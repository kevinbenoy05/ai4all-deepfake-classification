```python
from diffusers import DDPMPipeline
import torch

pipeline = DDPMPipeline.from_pretrained(
    "Tomisin05/unconditional-anime-diffusion-model",
    torch_dtype=torch.float16
).to("cuda")
```

```python
num_images_to_generate = 80  # @param {type:"integer"} You can change this value to generate more or fewer images.

# script to generate x number of images
```

```python
images = pipeline(
    num_inference_steps=400,
    batch_size=num_images_to_generate,
    output_type="pil"
).images
```

```python
# show the generated image
import matplotlib.pyplot as plt

# resize all images and increase resolution
for i in range(len(images)):
    # images[i] = images[i].resize((64, 64))
    # images[i] = images[i].resize((512, 512))
    images[i] = images[i].resize((4096, 4096))


# show all images
for i in range(len(images)):
    plt.figure(figsize=(50, 50))
    plt.subplot(1, len(images), i + 1)
    plt.imshow(images[i])
    plt.axis('off')
    plt.show()
# plt.imshow(images[78]) # Display the first generated image
# plt.axis('off') # Hide axes for cleaner image display
# plt.show()
```
