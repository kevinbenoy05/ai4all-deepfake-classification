```python
from datasets import load_dataset

ds = load_dataset("Tomisin05/generated-anime-images")
```

```python
# view random 20 images from ds
import random
import matplotlib.pyplot as plt

print(len(ds["train"]))

def view_random_images(ds, num_images=20):
    for i in range(num_images):
        plt.figure(figsize=(50, 50))
        plt.subplot(1, num_images, i + 1)
        plt.imshow(ds[random.randint(0, len(ds))]["image"])
        plt.axis('off')
        plt.show()

view_random_images(ds["train"])


```
